## 누락 데이터 처리 II

```python
'''
4-38
판다스의 누락데이터 처리 = NaN
dropna()
dropna(how='all')          행, 열이 모두 NaN 값일 때만
dropna(how='any')          행, 열에 하나라도 NaN 값 있을 때   ,   (default)
dropna(thresh=3)           특정 갯수 이상의 (비누락) 데이터가 있는 행만 남긴다.

fillna                     누락데이터 값을 대체
isnull()
isna()                     isnull()
notnull()
'''
```

```python
from numpy import nan as NA

df = DataFrame([[1,6,5,4],[1,NA,NA],[NA],[NA,6,5,3]])
df
```

```python
removeNA = df.dropna()
removeNA
```

```python
removeNA = df.dropna(how='all')
removeNA
```

```python
removeNA = df.dropna(thresh=1)
removeNA
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean())
# df.fillna(0)
df.fillna(value=23)
```
