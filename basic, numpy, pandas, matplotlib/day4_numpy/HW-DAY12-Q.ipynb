# 1. 연봉 계산하기

### 철수는 시급 15000원을 받으며 회사를 다니고 있습니다.<br>하루마다 일하는 시간은 5일을 간격으로 반복되고, 매 달마다 일하는 날 수가 정해져 있습니다.<br>연말에 매달 월급의 10%의 합만큼 보너스를 받는다고 했을 때, 철수의 매 달의 월급과 연봉은 각각 얼마인가요?
<br> correction. 8,8,8,7,7 이 리셋되어 매달 반복

```python
# hours - 5일을 간격으로 반복되는 근무시간, days - 달마다 일하는 일 수(1-12월)
hours = [8,8,8,7,7]
days = [23,18,24,23,24,22,23,22,23,22,20,10]
```

```python
# 코드를 작성해주세요.
money_for_day = 15000

month_money_list = list()

for idx, d in enumerate(days):
    month_money_list.append ( ( days[idx]//5 ) * sum(hours) )
    month_money_list[idx] += sum( hours[:days[idx]%5] )
    month_money_list[idx] *= (15000)
    
for i, m in enumerate(month_money_list):
    print('%d 월의 월급은 %.1f만원입니다.'%(i+1, m//10000))
print('\n총 연봉은 ', sum(month_money_list)*(1.1)//10000,'만원 입니다.')
```
