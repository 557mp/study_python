## 데이터 병합 - cancat, merge

```python
'''
concat - 같은 col 가지는 경우, default(위/아래) 좌/우도 가능
merge - 공통의 col을 기준으로 병합, default(이너 조인)
      - on : 공통 col
      - how : 조인 방식
'''

# concat

df1 = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

df2 = DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})

df2
```

```python
# result = pd.concat(df1, df2)
result = pd.concat([df1,df2], ignore_index=True)
result
```

```python
result = pd.concat([df1,df2], axis=1)
result
```

```python
result = pd.concat([df1,df2], keys=['X','Y'])
result
```

```python
# ㅡㄷ

'''
df1,df2는 인덱스만 다를뿐 모든 값이 동일함
df3는 Color_num이라는 컬럼이 다르다. 값도 다르다.

'''
df1 = DataFrame({ 'Year':[2001,2002,2003,2004],
                  'Product_Code':[11,22,33,44],
                  'Price':[10000,20000,30000,40000]},
                   index=list('1234'))
df1
```

```python
df2 = DataFrame({ 'Year':[2001,2002,2003,2004],
                  'Product_Code':[11,22,33,44],
                  'Price':[10000,20000,30000,40000]},
                   index=list('5678'))
df2
```

```python
df3 = DataFrame({ 'Year':[2001,2003,2004,2005],
                  'Product_Code':[11,22,33,44],
                  'Color_num':[33,44,55,99]},
                   index=list('1234'))
df3
```

```python
# 1. df1, df2  Merge

result1 = pd.merge(df1, df2)
result1
```

```python
# 2. 공톨 칼럼 기준으로 Merge

result2 = pd.merge(df1,df3, on='Year')
result2
```

```python
result3 = pd.merge(df1,df3, on=['Year', 'Product_Code'])
result3
```

```python
# set_indeX()

result1.set_index('Year', inplace=True)
result1
```

```python
# 3. Merge 방식 : how()  inner, outer

result5 = pd.merge(df1, df3, on='Year', how='outer')
result5
```

```python
# 3. Merge 방식 : how()  inner, outer

result7 = pd.merge(df1, df3, on='Year', how='right')
result7
```

## 실전 데이터 실습

```python
tipdf = pd.read_csv('../data/tips.csv')
tipdf.head()
```

```python
# 1. 날짜별로 Groupby(), sum, count

tipdf.groupby(['day']).count()
```

```python
'''
groupy() 결과를 새로운 칼럼에 할당
새로운 칼럼에 tip에 대한 평균값을 넣어보자
'''
tipgroup=tipdf.groupby(['day'])
tipgroup['tip'].mean().sort_values(ascending=False)
tipgroup[['tip']].mean().sort_values(['tip'], ascending=False).plot(kind='bar', grid=True)
plt.show()
```

```python
# 2. 성별, 흡연여부별 그룹핑하여 평균 통계함수 적용
tipgroup2 = tipdf.groupby(['sex', 'smoker']).agg('mean')             # 'day', 'time' 값은 왜 안뜰까??   통계함수는 nurmeric only !
tipgroup2
```

```python
tipgroup2.plot(kind='bar', grid=True)
plt.xticks(rotation=0)
plt.show()
```

## Pivot Tables

```python
data
```

```python
'''
엑셀의 피봇테이블과 유사
Data Frame 데이터를 reshape 한다.
pivot(), pivot_table()
'''
df1=DataFrame(data)
df1
```

```python
df1.pivot_table('인구', index='도시', columns=['연도'], aggfunc='sum')   ## 키워드 매개변수, 위치매개변수 ( values='인구' )
```

```python
df1.pivot_table(values=['인구'], index=['도시'], columns=['연도'], margins=True)     # 합, 전체 데이터의 평균. 평균의 평균
```

```python
df1['인구'].mean()
```

```python
# 이중 groupby과 비슷한 효과

df1.pivot_table(values=['인구'], index=['도시', '연도'])    # ascending = True (디폴트)
```
