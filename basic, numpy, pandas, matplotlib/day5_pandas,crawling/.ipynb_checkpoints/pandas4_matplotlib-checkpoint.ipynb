## 겹쳐진 그래프

```python
import numpy as np
```

```python
# 0 부터 5 미만까지 값 받기. 0,2 간격

t=np.arange(0,5,0.2)
t
```

```python
'''
t 값들을 x축, y축 값으로 입력
'''
plt.plot(t,t, 'g--')
plt.show()
```

```python
plt.plot(t,t, 'g--', t,t**2, 'ys')
plt.show()
```

```python
plt.plot(t,t, 'g--', t,t**2, 'ys', t, t*2, 'g^')
plt.show()
```

```python
plt.plot(t,t,'r--', label='t')
plt.plot(t,t**2,'g^', label='t**2')
plt.plot(t,t*2,'ys', label='t*2')
plt.legend()      # LABEL 나타내기 위해서만
plt.title('legend() for label')
plt.xlabel('xxxxx')
plt.ylabel('YYY')
plt.yticks(rotation=270)
plt.grid()
plt.show()
```

## 파도 모야의 그래프

```python
# sin(), cos()

t = np.arange(0,12,0.01)
plt.figure(figsize=(12,2))      # figure 설정은 plot 전에 해야한다.
plt.plot(t, np.sin(t), 'r', label='S i n', lw=5)     # lw : Line Width
plt.plot(t, np.cos(t), 'g', label='C o s', lw=1)
plt.title('S & C')
plt.legend()
plt.xlabel('x label')
plt.ylabel('y label')
plt.show()
```

## Pandas를 이용한 Visualization

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0,50,50),
        'd': np.random.randn(50)
       }

data
```

```python
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
data
```

```python
import pandas as pd
from pandas import DataFrame, Series

data1 = DataFrame(data)
data1.tail()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data1)         # x축 : 'a'   y축 : 'c'   data = data1 
```

```python
'''
x축: 1/1/2000 부터 1000 일간
y축: 누적되는 값 (음수도 포함)
'''
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000)) #period default = days
ts
```

```python
ts.plot()
```

```python
ts.plot(kind='bar')
```

## Box Plot

```python
import seaborn as sns
import numpy as np

# 0부터 100까지 총 101개의 숫자를 만든다.
xs = np.array(np.linspace(start=0, stop=100, num=101))        # lin space
xs
```

```python
df = DataFrame(xs, columns=['My Numbers'])
df
```

```python
plt.figure(figsize=(6,6))
df.boxplot(column=['My Numbers'])
plt.yticks(np.arange(0,101,step=5))
plt.show()                                 # 박스 중간값( 50% ), 25%, 75%,  사분위
```

## 실전 데이터

```python
# tips 데이터 이용, Seaborn   boxplot 그리자
#tips  error
tips = sns.load_dataset('tips')             # 라이브러리에서 제공
tips
```

```python
plt.figure(figsize=(6,4))
sns.boxenplot(x=tips['total_bill'])
plt.show()
```

```python
# tips.sort_values(by=['total_bill'], ascending=False).head()
tips.sort_values(by=['total_bill'], ascending=False).tail()
```

```python
# 날짜별 지불금액 분석
'''
x축 : 날짜
y축 : 'total_value'
'''
sns.boxplot(x=tips['day'], y=tips['total_bill'], data=tips)
plt.show()
```

```python
# 요일별 지출 총액, 흡연에 따른 비교

sns.boxplot(x=tips['day'], y=tips['total_bill'], hue='smoker', data=tips)
plt.show()
```

```python
# 선, 쉐도우

sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips)
plt.show()
```

## Heat Map

```python
flights=sns.load_dataset('flights')
flights.info()
```

```python
# pivot 테이블 사용.
# index = month, value = passengers, col = year

flights=flights.pivot_table(index='month', values='passengers', columns='year')
flights
```

```python
'''
x축: columns
y축: index
값: passengers
'''
plt.figure(figsize=(8,6))
sns.heatmap(data=flights, annot=True, fmt='d')  # annotaion, decimal
plt.xticks(rotation=0)
plt.show()
```
